```python
# specific the column name that I like to see as:
new_column_dict= {
    'timestamp':'time',
    'cnt':'count',
    't1': 'temp_real_C',
    't2':'temp_feels_like_C',
    'hum': 'humidity_percentage',
    'wind_speed':'wind_speed_kmph',
    'weather_code': 'weather',
    'is_holiday': 'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
    }

```

```python
# Rename the headers of the table to my specific name:
bikes.rename(new_column_dict, axis=1 ,inplace=True)
```

```python
bikes
```

```python
# converting the humidity values to actual percebtage values :
bikes.humidity_percentage= bikes.humidity_percentage/100
```

```python
# creating the season dictories so, that wev can map the intergers 0-3 to the actual written values: 
season_dict = {
    '0.0': 'spring',
    '1.0':  'summer',
    '3.0':  'autumn',
    '2.0':  'winter'
}

season_dict
```

```python
weather_dict = {
    '1.0': 'clear',
    '2.0': 'scattered',
    '3.0': 'broken_clouds',
    '4.0': 'cloudy',
    '7.0':' rain',
    '26.0': 'snowfall',
    '10.0':  'thunderstorm'
}

weather_dict
```

```python
# changing the seasons column data type to string:
bikes.season =bikes.season.astype('str')

# changing the queries or the values that we wrote above to some other values :

bikes.weather= bikes.weather.astype('str')
```

```python
# implementing the values that changed for the season and weather into main csv file:
bikes.season= bikes.season.map(season_dict)

bikes.weather= bikes.weather.map(weather_dict)
```
